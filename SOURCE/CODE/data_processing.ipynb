```python
# ✅ Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBRegressor
from sklearn.linear_model import Ridge
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, LSTM, LayerNormalization, MultiHeadAttention, Add
from tensorflow.keras.optimizers import Adam

# ✅ Load Dataset
df = pd.read_csv("/Users/disastershubz/Downloads/MacroTrends_Data_Download_TSLA.csv", skiprows=14)
df.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.sort_index(inplace=True)

# ✅ Normalize Data
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[['close', 'volume']])
df_scaled = pd.DataFrame(df_scaled, index=df.index, columns=['close', 'volume'])

# ✅ Train-Test Split
train_size = int(len(df_scaled) * 0.8)
train, test = df_scaled.iloc[:train_size], df_scaled.iloc[train_size:]

# ✅ Prepare Sequences for LSTM & Transformer
def create_sequences(data, time_steps=60):
    X, y = [], []
    for i in range(time_steps, len(data)):
        X.append(data.iloc[i-time_steps:i].values)
        y.append(data['close'].iloc[i])
    return np.array(X), np.array(y)

time_steps = 60
X_train, y_train = create_sequences(train, time_steps)
X_test, y_test = create_sequences(test, time_steps)
test = test.iloc[time_steps:]

# ✅ Define Transformer Model
def build_transformer_model(input_shape):
    inputs = Input(shape=input_shape)
    
    # Multi-Head Attention
    attn_output = MultiHeadAttention(num_heads=4, key_dim=64)(inputs, inputs)
    attn_output = Add()([attn_output, inputs])  # Residual Connection
    attn_output = LayerNormalization()(attn_output)

    # Feed Forward Network
    ffn_output = Dense(128, activation='relu')(attn_output)
    ffn_output = Dropout(0.2)(ffn_output)
    ffn_output = Dense(64, activation='relu')(ffn_output)
    ffn_output = Dense(1)(ffn_output)
    
    model = Model(inputs, ffn_output)
    model.compile(optimizer=Adam(learning_rate=0.0005), loss='mse')
    return model

# ✅ Train Transformer Model
transformer_model = build_transformer_model((X_train.shape[1], X_train.shape[2]))
transformer_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
transformer_pred = transformer_model.predict(X_test).flatten()

# ✅ Train LSTM Model
lstm_model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(50),
    Dense(1)
])
lstm_model.compile(optimizer=Adam(learning_rate=0.0005), loss='mse')
lstm_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

# ✅ Predict LSTM
lstm_pred = lstm_model.predict(X_test).flatten()

# ✅ XGBoost Model
train_xgb = train[['close', 'volume']]
test_xgb = test[['close', 'volume']]
xgb_model = XGBRegressor(n_estimators=200, max_depth=5, learning_rate=0.05, subsample=0.9)
xgb_model.fit(train_xgb, train['close'])
xgb_pred = xgb_model.predict(test_xgb)

# ✅ Align Predictions (Trim to Minimum Length)
min_len = min(len(lstm_pred), len(xgb_pred), len(transformer_pred))
lstm_pred, xgb_pred, transformer_pred = lstm_pred[:min_len], xgb_pred[:min_len], transformer_pred[:min_len]

# ✅ Rescale Predictions Back to Original Prices
dummy_array = np.zeros((min_len, df_scaled.shape[1]))

dummy_array[:, 0] = lstm_pred.flatten()
lstm_fixed = scaler.inverse_transform(dummy_array)[:, 0]

dummy_array[:, 0] = xgb_pred.flatten()
xgb_fixed = scaler.inverse_transform(dummy_array)[:, 0]

dummy_array[:, 0] = transformer_pred.flatten()
transformer_fixed = scaler.inverse_transform(dummy_array)[:, 0]

# ✅ Train Ridge Regression Meta-Model (Stacking)
meta_X = np.column_stack((lstm_fixed, xgb_fixed, transformer_fixed))
meta_y = df['close'].iloc[-min_len:].values

meta_model = Ridge(alpha=0.5)
meta_model.fit(meta_X, meta_y)

# ✅ Final Stacked Model Prediction
final_pred = meta_model.predict(meta_X)

# ✅ Apply Rolling Average for Additional Smoothing
final_pred_smooth = pd.Series(final_pred).rolling(window=5).mean()

# ✅ Generate Future Dates
future_dates = pd.date_range(start=test.index[-1], periods=len(final_pred_smooth) + 1, freq='D')[1:]

# ✅ Plot Results
plt.figure(figsize=(12, 6))
plt.plot(df.index, df["close"], label="Historical Prices", color="black")
plt.plot(future_dates[:len(final_pred_smooth)], final_pred_smooth, label="Hybrid Transformer + LSTM + XGBoost (Stacked, Final)", linestyle="dashed", color="blue")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.title("Stock Price Forecast Using Hybrid Transformer + LSTM + XGBoost")
plt.legend()
plt.grid(True)
plt.show()
```
